```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import LinearSVC as SKLearnSVC
from picolearn.svm import SVC as PicoLearnSVC
```

```python
def plot_decision_boundary(X, y, w, b):
    plt.scatter(X[:, 0][:2], X[:,1][:2], color="blue", marker="o")
    plt.scatter(X[:, 0][:2], X[:,1][2:], color="green", marker="x")
    x = np.linspace(-5, 5, 100)
    y_0 = w[0][0] * x + b 
    y_1 = w[0][1] * x + b 
    plt.plot(x, y_0, linestyle='--', color='red', linewidth=2) 
    plt.plot(x, y_1, linestyle='--', color='red', linewidth=2) 
    plt.plot()
    plt.show()
```

```python
X = np.array([[-1, -1], [-2, -1], [1, 1], [2, 1]])
y = np.array([1, 1, 2, 2])
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y = np.array([0, 1, 0, 1])
X_test = np.array([[-0.8, -1]])
```

```python
X[:, 0]
X[:, 1]
```

```python

sk_model = SKLearnSVC()
sk_model.fit(X, y)

pl_model = PicoLearnSVC()
pl_model.fit(X, y)
```

```python
plot_decision_boundary(X,y, sk_model.coef_, sk_model.intercept_)
```

```python
plot_decision_boundary(X,y, pl_model.coef_, pl_model.intercept_)
```
